```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image
```

```python
apple_pie = []
for i in range(1,11):
    img = Image.open('./data/food-101/images/apple_pie/' + os.listdir('./data/food-101/images/apple_pie')[i])
    apple_pie.append(img)

# plot the first 10 apple_pie images
fig = plt.figure(figsize=(20, 10))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(apple_pie[i])
    plt.axis('off')
plt.show()
```
